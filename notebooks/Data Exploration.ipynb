Looking at the relation between a StockCode its Description, it was found that any given StockCode should have a fixed Description. When looked at the occurance of a Description under any StockCode, it can be seen that there existed multiple descriptions under a StockCode. These extra Description will be replaced by the most repeated one under the same StockCode

```python
# Step 1: Get the most frequent description for each StockCode
def get_most_frequent_description(group):
    # Get the mode (most frequent value) in the Description column
    mode_values = group.mode()['Description']
    # If there are multiple modes (a tie), choose the first one
    return mode_values.iloc[0] if not mode_values.empty else None

# Group by StockCode and apply the function
most_frequent_desc = df.groupby('StockCode').apply(get_most_frequent_description).reset_index(name='Most_Frequent_Description')

# Step 2: Merge the most frequent description back into the original dataframe
df = df.merge(most_frequent_desc, on='StockCode', how='left')

# Step 3: Replace the Description column with the most frequent one
df['Description'] = df['Most_Frequent_Description']

# Drop the extra column used for merging
df.drop(columns=['Most_Frequent_Description'], inplace=True)
```

```python
import pyodbc

def create_table_from_dict(table_name, columns, conn, drop_if_exists=False, skip_if_exists=True):
    cursor = conn.cursor()

    if skip_if_exists:
        # Skip if table exists
        cursor.execute(f"IF OBJECT_ID('{table_name}', 'U') IS NULL BEGIN CREATE TABLE {table_name} ({', '.join([f'{col} {dtype}' for col, dtype in columns.items()])}) END")
        conn.commit()
        print(f"Table {table_name} created (if not exists)")

    elif drop_if_exists:
        # Drop the table if it exists and recreate it
        cursor.execute(f"IF OBJECT_ID('{table_name}', 'U') IS NOT NULL DROP TABLE {table_name}")
        conn.commit()
        column_defs = ", ".join([f"{col} {dtype}" for col, dtype in columns.items()])
        cursor.execute(f"CREATE TABLE {table_name} ({column_defs})")
        conn.commit()
        print(f"Table {table_name} dropped (if exists) and recreated")

    else:
        try:
            # Try creating the table
            column_defs = ", ".join([f"{col} {dtype}" for col, dtype in columns.items()])
            cursor.execute(f"CREATE TABLE {table_name} ({column_defs})")
            conn.commit()
            print(f"Table {table_name} created successfully!")
        except pyodbc.OperationalError as e:
            print(f" Error: {e}")


# Define the columns as a dictionary
columns = {
    "InvoiceNo": "VARCHAR(20)",
    "StockCode": "VARCHAR(20)",
    "Description": "NVARCHAR(255)",
    "Quantity": "INT",
    "InvoiceDate": "DATETIME",
    "UnitPrice": "DECIMAL(10, 2)",
    "CustomerID": "INT",
    "Country": "NVARCHAR(100)",
    "TotalPrice": "DECIMAL(10, 2)",
    "IsReturn": "BIT"
}

# Connect to SQL Server
conn = pyodbc.connect(
    "DRIVER={SQL Server};"
    "SERVER=DataSpace\SQLEXPRESS;"  # Adjust your server here
    "DATABASE=Test;"  # Adjust your database name here
    "Trusted_Connection=yes;"
)

# Create the table dynamically
create_table_from_dict("SalesData", columns, conn, drop_if_exists=False, skip_if_exists=True)

```

```python
def insert_dataframe_to_sql(df, table_name, conn):
    cursor = conn.cursor()
    placeholders = ', '.join(['?' for _ in df.columns])
    columns = ', '.join(df.columns)
    insert_sql = f"INSERT INTO {table_name} ({columns}) VALUES ({placeholders})"

    df = df.where(pd.notnull(df), None)
    data = [tuple(x) for x in df.to_numpy()]

    cursor.executemany(insert_sql, data)
    conn.commit()
    cursor.close()

    print(f"Bulk-inserted {len(data)} rows into '{table_name}'")

```

```python
insert_dataframe_to_sql(df, "SalesData", conn)
```
